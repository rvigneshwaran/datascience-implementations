```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a sample DataFrame
data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Sales': [1000, 1200, 1500, 1800, 2000],
    'Expenses': [800, 900, 1000, 1100, 1200]
}
df = pd.DataFrame(data)
```

```python
# Example 1: Line plot of Sales over the Years
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the Years')
plt.grid(True)
plt.show()
```

```python
# Example 2: Line plot with multiple data series (Sales and Expenses)
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'], label='Sales')
plt.plot(df['Year'], df['Expenses'], label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Sales and Expenses over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 3: Scatter plot of Sales against Expenses
plt.figure(figsize=(6, 6))
plt.scatter(df['Expenses'], df['Sales'])
plt.xlabel('Expenses')
plt.ylabel('Sales')
plt.title('Sales vs. Expenses')
plt.grid(True)
plt.show()
```

```python
# Example 4: Bar plot of Sales over the Years
plt.figure(figsize=(8, 4))
plt.bar(df['Year'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the Years')
plt.grid(True)
plt.show()
```

```python
# Example 5: Horizontal bar plot of Sales over the Years
plt.figure(figsize=(8, 4))
plt.barh(df['Year'], df['Sales'])
plt.xlabel('Sales')
plt.ylabel('Year')
plt.title('Sales over the Years')
plt.grid(True)
plt.show()
```

```python
# Example 6: Stacked bar plot of Sales and Expenses over the Years
plt.figure(figsize=(8, 4))
plt.bar(df['Year'], df['Sales'], label='Sales')
plt.bar(df['Year'], df['Expenses'], bottom=df['Sales'], label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Stacked Sales and Expenses over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 7: Pie chart of Sales distribution
plt.figure(figsize=(6, 6))
plt.pie(df['Sales'], labels=df['Year'], autopct='%1.1f%%')
plt.title('Sales distribution')
plt.show()
```

```python
# Example 8: Box plot of Sales and Expenses
plt.figure(figsize=(6, 4))
plt.boxplot([df['Sales'], df['Expenses']], labels=['Sales', 'Expenses'])
plt.ylabel('Amount')
plt.title('Sales and Expenses Box Plot')
plt.grid(True)
plt.show()
```

```python
# Example 9: Histogram of Sales
plt.figure(figsize=(6, 4))
plt.hist(df['Sales'], bins=10, edgecolor='black')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Sales Histogram')
plt.grid(True)
plt.show()
```

```python
# Example 10: Area plot of Sales and Expenses over the Years
plt.figure(figsize=(8, 4))
plt.stackplot(df['Year'], df['Sales'], df['Expenses'], labels=['Sales', 'Expenses'])
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Area plot of Sales and Expenses')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 11: KDE (Kernel Density Estimation) plot of Sales
plt.figure(figsize=(6, 4))
df['Sales'].plot.kde()
plt.xlabel('Sales')
plt.title('Kernel Density Estimation of Sales')
plt.grid(True)
plt.show()
```

```python
# Example 12: Hexbin plot of Sales against Expenses
plt.figure(figsize=(6, 6))
plt.hexbin(df['Expenses'], df['Sales'], gridsize=10, cmap='Blues')
plt.xlabel('Expenses')
plt.ylabel('Sales')
plt.title('Hexbin plot of Sales vs. Expenses')
plt.colorbar(label='Count')
plt.grid(True)
plt.show()
```

```python
# Example 13: Area plot of Sales over the Years
plt.figure(figsize=(8, 4))
plt.fill_between(df['Year'], df['Sales'], alpha=0.5)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Area plot of Sales over the Years')
plt.grid(True)
plt.show()
```

```python
# Example 14: Stacked area plot of Sales and Expenses over the Years
plt.figure(figsize=(8, 4))
plt.stackplot(df['Year'], df['Sales'], df['Expenses'], labels=['Sales', 'Expenses'], alpha=0.5)
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Stacked Area plot of Sales and Expenses over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Example 15: Dual-axis line plot of Sales and Expenses over the Years
fig, ax1 = plt.subplots(figsize=(8, 4))

ax2 = ax1.twinx()
ax1.plot(df['Year'], df['Sales'], 'g-')
ax2.plot(df['Year'], df['Expenses'], 'b-')

ax1.set_xlabel('Year')
ax1.set_ylabel('Sales', color='g')
ax2.set_ylabel('Expenses', color='b')
ax1.set_title('Dual-axis line plot of Sales and Expenses over the Years')
plt.grid(True)
plt.show()
```

```python
# Example 16: Line plot with markers
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'], marker='o')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the Years')
plt.grid(True)
plt.show()
```

```python
# Example 17: Line plot with different line styles
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'], linestyle='--', marker='o', color='r')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the Years')
plt.grid(True)
plt.show()
```

```python
# Example 18: Line plot with custom color map
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'], c=np.arange(len(df['Year'])), cmap='viridis', marker='o')
plt.colorbar(label='Index')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the Years with Custom Color Map')
plt.grid(True)
plt.show()
```

```python
# Example 19: Line plot with log scale on the y-axis
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'])
plt.yscale('log')
plt.xlabel('Year')
plt.ylabel('Sales (Log Scale)')
plt.title('Sales over the Years (Log Scale)')
plt.grid(True)
plt.show()
```

```python
# Example 20: Scatter plot with custom marker size and color
plt.figure(figsize=(6, 6))
plt.scatter(df['Expenses'], df['Sales'], s=df['Year']*10, c=df['Year'], cmap='plasma')
plt.xlabel('Expenses')
plt.ylabel('Sales')
plt.title('Sales vs. Expenses')
plt.colorbar(label='Year')
plt.grid(True)
plt.show()
```

```python
# Example 21: Bar plot with error bars
plt.figure(figsize=(8, 4))
plt.bar(df['Year'], df['Sales'], yerr=df['Sales']*0.1, capsize=5)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the Years with Error Bars')
plt.grid(True)
plt.show()
```

```python
# Example 22: Plotting multiple plots in one figure using subplots
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(df['Year'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the Years')

plt.subplot(2, 2, 2)
plt.scatter(df['Expenses'], df['Sales'])
plt.xlabel('Expenses')
plt.ylabel('Sales')
plt.title('Sales vs. Expenses')

plt.subplot(2, 2, 3)
plt.bar(df['Year'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the Years')

plt.subplot(2, 2, 4)
plt.pie(df['Sales'], labels=df['Year'], autopct='%1.1f%%')
plt.title('Sales distribution')

plt.tight_layout()
plt.show()
```

```python
# Example 23: Customizing plot appearance - color, style, and labels
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'], color='purple', linestyle='-.', marker='o', markersize=8, label='Sales')
plt.plot(df['Year'], df['Expenses'], color='orange', linestyle='--', marker='s', markersize=8, label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Customized Sales and Expenses over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 24: Adding annotations to the plot
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'], marker='o', label='Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the Years')
plt.annotate('Maximum Sales', xy=(2014, 2000), xytext=(2014.5, 1800),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 25: Creating a multi-line plot using a loop
plt.figure(figsize=(8, 4))
for col in ['Sales', 'Expenses']:
    plt.plot(df['Year'], df[col], marker='o', label=col)
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Multi-line plot of Sales and Expenses over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 26: Adding a legend outside the plot
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'], marker='o', label='Sales')
plt.plot(df['Year'], df['Expenses'], marker='o', label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Sales and Expenses over the Years')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

```python
# Example 27: Using style sheets to customize plots
plt.style.use('ggplot')
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'], marker='o', label='Sales')
plt.plot(df['Year'], df['Expenses'], marker='o', label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Sales and Expenses over the Years')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Example 28: Saving the plot to a file
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'], marker='o', label='Sales')
plt.plot(df['Year'], df['Expenses'], marker='o', label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Sales and Expenses over the Years')
plt.legend()
plt.grid(True)
plt.savefig('sales_and_expenses.png')
plt.show()
```

```python
# Example 29: Customizing the figure size and resolution
plt.figure(figsize=(12, 6), dpi=100)
plt.plot(df['Year'], df['Sales'], marker='o', label='Sales')
plt.plot(df['Year'], df['Expenses'], marker='o', label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Sales and Expenses over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 30: Creating a bar plot with date labels on the x-axis
df['Date'] = pd.to_datetime(df['Year'], format='%Y')
plt.figure(figsize=(8, 4))
plt.bar(df['Date'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the Years with Date Labels')
plt.grid(True)
plt.show()
```
