```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a sample DataFrame
data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Sales': [1000, 1200, 1500, 1800, 2000],
    'Expenses': [800, 900, 1000, 1100, 1200]
}
df = pd.DataFrame(data)
```

```python
# Example 1: Line plot of Sales over the Years
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the Years')
plt.grid(True)
plt.show()
```

```python
# Example 2: Line plot with multiple data series (Sales and Expenses)
plt.figure(figsize=(8, 4))
plt.plot(df['Year'], df['Sales'], label='Sales')
plt.plot(df['Year'], df['Expenses'], label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Sales and Expenses over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 3: Scatter plot of Sales against Expenses
plt.figure(figsize=(6, 6))
plt.scatter(df['Expenses'], df['Sales'])
plt.xlabel('Expenses')
plt.ylabel('Sales')
plt.title('Sales vs. Expenses')
plt.grid(True)
plt.show()
```

```python
# Example 4: Bar plot of Sales over the Years
plt.figure(figsize=(8, 4))
plt.bar(df['Year'], df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales over the Years')
plt.grid(True)
plt.show()
```

```python
# Example 5: Horizontal bar plot of Sales over the Years
plt.figure(figsize=(8, 4))
plt.barh(df['Year'], df['Sales'])
plt.xlabel('Sales')
plt.ylabel('Year')
plt.title('Sales over the Years')
plt.grid(True)
plt.show()
```
