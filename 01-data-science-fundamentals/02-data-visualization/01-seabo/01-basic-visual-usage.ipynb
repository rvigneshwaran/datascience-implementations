```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_data = sns.load_dataset("iris")
```

```python
iris_data.head()
```

```python
iris_data.columns
```

```python
plt.figure(figsize = (7, 3))
plt.title('Lineplot Using Seaborn')
sns.lineplot(x="sepal_length", y="sepal_width", data=iris_data)
sns.set_style("dark")
plt.show()
```
