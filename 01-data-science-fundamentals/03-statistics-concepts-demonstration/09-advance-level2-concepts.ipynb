This Python notebook covers the following concepts related to Random Variables:
1. What are random variables?
    - Example 1: Rolling a fair six-sided die and defining the random variable for the outcome.
2. Multivariate random variables:
    - Example 2: Simulating samples from a bivariate normal distribution.
3. Discrete random variables:
    - Example 3: Simulating samples from a binomial distribution (number of successes in coin tosses).
4. Continuous random variables:
    - Example 4: Simulating samples from a normal distribution.
5. Functions of random variables:
    - Example 5: Generating samples from a function of random variables (exponential distribution).
6. Creating random variables:
    - Example 6: Creating a custom random variable using numpy's random choice.

```python
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Setting the random seed for reproducibility
np.random.seed(42)
```

```python
# What are random variables?

# Example 1: Rolling a fair six-sided die
die_outcomes = [1, 2, 3, 4, 5, 6]
prob_die = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]
```

```python
# Multivariate Random Variables

# Example 2: Simulating bivariate normal distribution
mean = [0, 0]
covariance_matrix = np.array([[1, 0.5], [0.5, 1]])
num_samples = 1000
```

```python
# Generating samples from the bivariate normal distribution
samples = np.random.multivariate_normal(mean, covariance_matrix, num_samples)
```

```python
# Discrete Random Variables

# Example 3: Simulating a discrete random variable (number of successes in coin tosses)
num_trials = 10
p_heads = 0.5
```

```python
# Generating samples from a binomial distribution
binomial_samples = np.random.binomial(num_trials, p_heads, num_samples)
```

```python
# Plotting the histogram of binomial samples
plt.figure(figsize=(8, 4))
plt.hist(binomial_samples, bins=np.arange(num_trials+2)-0.5, density=True, alpha=0.75)
plt.title("Histogram of Binomial Samples")
plt.xlabel("Number of Successes (Heads)")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

```python
# Continuous Random Variables

# Example 4: Simulating a continuous random variable (normal distribution)
mean = 0
std_dev = 1
```

```python
# Generating samples from a normal distribution
normal_samples = np.random.normal(mean, std_dev, num_samples)
```

```python
# Plotting the histogram of normal samples
plt.figure(figsize=(8, 4))
plt.hist(normal_samples, bins=20, density=True, alpha=0.75)
plt.title("Histogram of Normal Samples")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

```python
# Functions of Random Variables

# Example 5: Generating samples from a function of random variables (exponential distribution)
lambda_ = 2

# Generating samples from an exponential distribution
exponential_samples = np.random.exponential(scale=1/lambda_, size=num_samples)
```

```python
# Plotting the histogram of exponential samples
plt.figure(figsize=(8, 4))
plt.hist(exponential_samples, bins=20, density=True, alpha=0.75)
plt.title("Histogram of Exponential Samples")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

```python
# Creating Random Variables

# Example 6: Creating a custom random variable using numpy's random choice
custom_outcomes = ["Red", "Green", "Blue"]
custom_probs = [0.4, 0.3, 0.3]
```

```python
# Generating samples from the custom random variable
custom_samples = np.random.choice(custom_outcomes, size=num_samples, p=custom_probs)
```

```python
# Plotting the histogram of custom samples
plt.figure(figsize=(6, 4))
sns.histplot(custom_samples, discrete=True)
plt.title("Histogram of Custom Random Variable Samples")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```
