```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load a pre-built dataset from Seaborn
df = sns.load_dataset("tips")
```

```python
# Example 1: Basic barplot
sns.barplot(data=df, x="day", y="total_bill")
plt.title("Basic Barplot")
plt.show()
```

```python
# Example 2: Barplot with hue
sns.barplot(data=df, x="day", y="total_bill", hue="sex")
plt.title("Barplot with Hue")
plt.show()
```

```python
# Example 3: Barplot with custom order
sns.barplot(data=df, x="day", y="total_bill", order=["Sun", "Sat", "Fri", "Thur"])
plt.title("Barplot with Custom Order")
plt.show()
```

```python
# Example 4: Barplot with custom palette
sns.barplot(data=df, x="day", y="total_bill", palette="Set2")
plt.title("Barplot with Custom Palette")
plt.show()
```

```python
# Example 5: Barplot with custom estimator
sns.barplot(data=df, x="day", y="total_bill", estimator=max)
plt.title("Barplot with Custom Estimator")
plt.show()
```

```python
# Example 6: Barplot with confidence intervals
sns.barplot(data=df, x="day", y="total_bill", ci=68)
plt.title("Barplot with Confidence Intervals")
plt.show()
```

```python
# Example 7: Barplot with error bars
sns.barplot(data=df, x="day", y="total_bill", ci="sd")
plt.title("Barplot with Error Bars")
plt.show()
```

```python
# Example 8: Barplot with capped error bars
sns.barplot(data=df, x="day", y="total_bill", ci="sd", capsize=0.2)
plt.title("Barplot with Capped Error Bars")
plt.show()
```

```python
# Example 9: Barplot with different cap widths
sns.barplot(data=df, x="day", y="total_bill", ci="sd", capsize=0.1)
plt.title("Barplot with Different Cap Widths")
plt.show()
```

```python
# Example 10: Barplot with custom palette and error bar colors
sns.barplot(data=df, x="day", y="total_bill", palette="Set2", ci="sd", errcolor="red")
plt.title("Barplot with Custom Palette and Error Bar Colors")
plt.show()
```

```python
# Example 11: Barplot with grouped bars
sns.barplot(data=df, x="day", y="total_bill", hue="sex", ci=None)
plt.title("Barplot with Grouped Bars")
plt.show()
```

```python
# Example 12: Barplot with stacked bars
sns.barplot(data=df, x="day", y="total_bill", hue="sex", ci=None, dodge=False)
plt.title("Barplot with Stacked Bars")
plt.show()
```

```python
# Example 13: Barplot with adjusted bar width
sns.barplot(data=df, x="day", y="total_bill", ci=None)
plt.title("Barplot with Adjusted Bar Width")
plt.xticks(rotation=45)
plt.show()
```

```python
# Example 14: Barplot with horizontal bars
sns.barplot(data=df, y="day", x="total_bill")
plt.title("Barplot with Horizontal Bars")
plt.show()
```

```python
# Example 15: Barplot with log scale
sns.barplot(data=df, x="day", y="total_bill")
plt.yscale("log")
plt.title("Barplot with Log Scale")
plt.show()
```
