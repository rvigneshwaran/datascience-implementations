```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load a pre-built dataset from Seaborn
iris = sns.load_dataset("iris")
```

```python
# Set the style for the plots
sns.set(style="whitegrid")
```

```python
# Example 1: Basic KDE Plot
sns.kdeplot(data=iris["sepal_length"])
plt.title("Basic KDE Plot")
plt.show()
```

```python
# Example 2: KDE Plot with Specific Color
sns.kdeplot(data=iris["sepal_length"], color="skyblue")
plt.title("KDE Plot with Specific Color")
plt.show()
```

```python
# Example 3: KDE Plot with Filled Area
sns.kdeplot(data=iris["sepal_length"], fill=True)
plt.title("KDE Plot with Filled Area")
plt.show()
```

```python
# Example 4: KDE Plot with Different Bandwidth
sns.kdeplot(data=iris["sepal_length"], bw_adjust=0.1)
plt.title("KDE Plot with Different Bandwidth")
plt.show()
```

```python
# Example 5: KDE Plot with Multiple Columns
sns.kdeplot(data=iris, x="sepal_length", hue="species", multiple="stack")
plt.title("KDE Plot with Multiple Columns")
plt.show()
```
