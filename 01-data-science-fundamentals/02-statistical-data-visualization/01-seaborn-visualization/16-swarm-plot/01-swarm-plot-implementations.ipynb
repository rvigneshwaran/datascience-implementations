```python
# Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load a pre-built dataset from Seaborn
iris = sns.load_dataset("iris")
```

```python
# Example 1: Basic swarmplot
sns.swarmplot(data=iris, x='species', y='sepal_length')
plt.title("Basic Swarmplot")
plt.show()
```

```python
# Example 2: Swarmplot with hue
sns.swarmplot(data=iris, x='species', y='sepal_length', hue='species')
plt.title("Swarmplot with Hue")
plt.legend(title="Species", loc='upper right')
plt.show()
```

```python
# Example 3: Swarmplot with vertical orientation
sns.swarmplot(data=iris, y='species', x='sepal_length', orient='h')
plt.title("Swarmplot with Vertical Orientation")
plt.show()
```

```python
# Example 4: Swarmplot with horizontal orientation and hue nesting
sns.swarmplot(data=iris, y='species', x='sepal_length', hue='species', orient='h', dodge=True)
plt.title("Swarmplot with Horizontal Orientation and Hue Nesting")
plt.show()
```

```python
# Example 5: Swarmplot with custom order
sns.swarmplot(data=iris, x='species', y='sepal_length', order=['virginica', 'setosa', 'versicolor'])
plt.title("Swarmplot with Custom Order")
plt.show()
```

```python
# Example 6: Swarmplot with custom hue order
sns.swarmplot(data=iris, x='species', y='sepal_length', hue='species', hue_order=['virginica', 'setosa', 'versicolor'])
plt.title("Swarmplot with Custom Hue Order")
plt.legend(title="Species", loc='upper right')
plt.show()
```

```python
# Example 7: Swarmplot with custom palette
sns.swarmplot(data=iris, x='species', y='sepal_length', palette='Set2')
plt.title("Swarmplot with Custom Palette")
plt.show()
```

```python
# Example 8: Swarmplot with marker customization
sns.swarmplot(data=iris, x='species', y='sepal_length', marker='D', markersize=8, markerfacecolor='red', markeredgecolor='black')
plt.title("Swarmplot with Marker Customization")
plt.show()
```

```python
# Example 9: Swarmplot with alpha blending
sns.swarmplot(data=iris, x='species', y='sepal_length', alpha=0.5)
plt.title("Swarmplot with Alpha Blending")
plt.show()
```

```python
# Example 10: Swarmplot with edgecolor and linewidth
sns.swarmplot(data=iris, x='species', y='sepal_length', edgecolor='black', linewidth=1)
plt.title("Swarmplot with Edgecolor and Linewidth")
plt.show()
```
