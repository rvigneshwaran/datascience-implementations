```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load a pre-built dataset from Seaborn
iris = sns.load_dataset("iris")
```

```python
# Set the style for the plots
sns.set(style="whitegrid")
```

```python
# Example 1: Basic Histogram
sns.histplot(data=iris, x="sepal_length")
plt.title("Basic Histogram")
plt.show()
```

```python
# Example 2: Histogram with Custom Number of Bins
sns.histplot(data=iris, x="sepal_length", bins=20)
plt.title("Histogram with Custom Number of Bins")
plt.show()
```

```python
# Example 3: Histogram with Specific Color
sns.histplot(data=iris, x="sepal_length", color="skyblue")
plt.title("Histogram with Specific Color")
plt.show()
```

```python
# Example 4: Histogram with Edge Color
sns.histplot(data=iris, x="sepal_length", edgecolor="black")
plt.title("Histogram with Edge Color")
plt.show()
```

```python
# Example 5: Histogram with Density Plot
sns.histplot(data=iris, x="sepal_length", kde=True)
plt.title("Histogram with Density Plot")
plt.show()
```

```python
# Example 6: Histogram with Rug Plot
sns.histplot(data=iris, x="sepal_length", kde=True, rug=True)
plt.title("Histogram with Rug Plot")
plt.show()
```

```python
# Example 7: Histogram with Multiple Columns
sns.histplot(data=iris, x="sepal_length", hue="species")
plt.title("Histogram with Multiple Columns")
plt.show()
```

```python
# Example 8: Histogram with Stacked Bars
sns.histplot(data=iris, x="sepal_length", hue="species", multiple="stack")
plt.title("Histogram with Stacked Bars")
plt.show()
```

```python
# Example 9: Histogram with Dodge Bars
sns.histplot(data=iris, x="sepal_length", hue="species", multiple="dodge")
plt.title("Histogram with Dodge Bars")
plt.show()
```

```python
# Example 10: Histogram with Cumulative Distribution
sns.histplot(data=iris, x="sepal_length", cumulative=True)
plt.title("Histogram with Cumulative Distribution")
plt.show()
```

```python
# Example 11: Histogram with Cumulative Density Function
sns.histplot(data=iris, x="sepal_length", cumulative=True, stat="density")
plt.title("Histogram with Cumulative Density Function")
plt.show()
```

```python
# Example 12: Histogram with Shaded Area
sns.histplot(data=iris, x="sepal_length", kde=True, kde_kws={"shade": True})
plt.title("Histogram with Shaded Area")
plt.show()
```

```python
# Example 13: Histogram with Log-scaled Axis
sns.histplot(data=iris, x="sepal_length", log_scale=True)
plt.title("Histogram with Log-scaled Axis")
plt.show()
```

```python
# Example 14: Histogram with Specific Axis Limits
sns.histplot(data=iris, x="sepal_length")
plt.xlim(4, 8)
plt.title("Histogram with Specific Axis Limits")
plt.show()
```

```python
# Example 15: Histogram with Custom Tick Positions
sns.histplot(data=iris, x="sepal_length")
plt.xticks([4, 5, 6, 7, 8])
plt.title("Histogram with Custom Tick Positions")
plt.show()
```
