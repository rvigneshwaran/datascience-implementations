```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_data = sns.load_dataset("iris")
```

```python
iris_data.head()
```

```python
iris_data.columns
```

```python
plt.figure(figsize = (7, 3))
plt.title('Lineplot Using Seaborn')
sns.lineplot(x="sepal_length", y="sepal_width", data=iris_data)
sns.set_style("dark") # Multiple Styles darkgrid,whitegrid,dark,white,ticks
sns.set_context("paper")
plt.show()
```

```python
# Setting Styles temprovarily and draw 2 Plots

def draw_line_plot():
    sns.lineplot(x="sepal_length", y="sepal_width", data=iris_data)

with sns.axes_style('darkgrid'):
    plt.subplot(211)
    draw_line_plot()
plt.subplot(212)
draw_line_plot()
```

```python
# Demonstrating Color Palette
palette = sns.color_palette()
sns.palplot(palette)
plt.show()
```

```python
# Diverging Color Palette
palette = sns.color_palette('PiYG', 11)
sns.palplot(palette) 
plt.show()
```

```python
# Sequential Color Palette
palette = sns.color_palette('Greens', 11)
sns.palplot(palette)
plt.show()
```

```python
# Set Color palette and Popeulate the Visual

# Setting "vlag" Palette
sns.set_palette('vlag')
plt.subplot(211)
draw_line_plot()
# Setting "Accent" Palette
sns.set_palette('Accent')
plt.subplot(212)
draw_line_plot()
plt.show()
```
