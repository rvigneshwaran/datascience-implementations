```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load a pre-built dataset from Seaborn
tips = sns.load_dataset("tips")
```

```python
tips.describe()
```

```python
tips.head(10)
```

```python
# Example 1: Basic countplot
sns.countplot(data=tips, x='day')
plt.title("Basic Countplot")
plt.show()
```

```python
# Example 2: Countplot with hue
sns.countplot(data=tips, x='day', hue='sex')
plt.title("Countplot with Hue")
plt.show()
```

```python
# Example 3: Countplot with custom order
sns.countplot(data=tips, x='day', order=['Sun', 'Sat', 'Thur', 'Fri'])
plt.title("Countplot with Custom Order")
plt.show()
```

```python
# Example 4: Countplot with custom palette
sns.countplot(data=tips, x='day', palette='husl')
plt.title("Countplot with Custom Palette")
plt.show()
```

```python
# Example 5: Countplot with hue and custom palette
sns.countplot(data=tips, x='day', hue='sex', palette='Set1')
plt.title("Countplot with Hue and Custom Palette")
plt.show()
```

```python
# Example 6: Countplot with hue and custom order
sns.countplot(data=tips, x='day', hue='sex', order=['Sun', 'Sat', 'Thur', 'Fri'])
plt.title("Countplot with Hue and Custom Order")
plt.show()
```

```python
# Example 7: Countplot with hue and custom hue order
sns.countplot(data=tips, x='day', hue='sex', hue_order=['Female', 'Male'])
plt.title("Countplot with Hue and Custom Hue Order")
plt.show()
```

```python
# Example 8: Countplot with hue and stacked bars
sns.countplot(data=tips, x='day', hue='sex', dodge=False)
plt.title("Countplot with Hue and Stacked Bars")
plt.show()
```

```python
# Example 9: Countplot with hue and adjusted bar width
sns.countplot(data=tips, x='day', hue='sex')
plt.title("Countplot with Hue and Adjusted Bar Width")
plt.xticks(rotation=45)
plt.show()
```

```python
# Example 10: Countplot with hue and horizontal bars
sns.countplot(data=tips, y='day', hue='sex')
plt.title("Countplot with Hue and Horizontal Bars")
plt.show()
```
