```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load a pre-built dataset from Seaborn
iris = sns.load_dataset("iris")
```

```python
# Set the style for the plots
sns.set(style="whitegrid")
```

```python
# Create a figure with subplots
fig, axes = plt.subplots(4, 4, figsize=(16, 16))
```

```python
# Example 1: Basic Rug Plot
sns.rugplot(data=iris, x="sepal_length", ax=axes[0, 0])
axes[0, 0].set_title("Basic Rug Plot")
```

```python
# Example 2: Rug Plot with Custom Color
sns.rugplot(data=iris, x="sepal_length", color="skyblue", ax=axes[0, 1])
axes[0, 1].set_title("Rug Plot with Custom Color")
```

```python
# Example 3: Horizontal Rug Plot
sns.rugplot(data=iris, y="sepal_length", vertical=True, ax=axes[0, 2])
axes[0, 2].set_title("Horizontal Rug Plot")
# Adjust layout
plt.tight_layout()
# Show the plots
plt.show()
```

```python
# Example 4: Rug Plot with Custom Height
sns.rugplot(data=iris, x="sepal_length", height=0.1, ax=axes[0, 3])
axes[0, 3].set_title("Rug Plot with Custom Height")
```

```python
# Example 5: Rug Plot with Custom Hue
sns.rugplot(data=iris, x="sepal_length", hue=iris["species"], palette="Set1", ax=axes[1, 0])
axes[1, 0].set_title("Rug Plot with Custom Hue")
```
