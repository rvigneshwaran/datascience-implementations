```python
# Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load a pre-built dataset from Seaborn
tips = sns.load_dataset("tips")
```

```python
# Example 1: Basic boxplot
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title("Basic Boxplot")
plt.show()
```

```python
# Example 2: Boxplot with horizontal orientation
sns.boxplot(data=tips, x='total_bill', y='day', orient='h')
plt.title("Boxplot with Horizontal Orientation")
plt.show()
```

```python
# Example 3: Boxplot with hue
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title("Boxplot with Hue")
plt.show()
```

```python
# Example 4: Boxplot with custom color
sns.boxplot(data=tips, x='day', y='total_bill', color='green')
plt.title("Boxplot with Custom Color")
plt.show()
```

```python
# Example 5: Boxplot with custom palette
sns.boxplot(data=tips, x='day', y='total_bill', palette='Set2')
plt.title("Boxplot with Custom Palette")
plt.show()
```

```python
# Example 6: Boxplot with notches
sns.boxplot(data=tips, x='day', y='total_bill', notch=True)
plt.title("Boxplot with Notches")
plt.show()
```

```python
# Example 7: Boxplot with quartile range specified
sns.boxplot(data=tips, x='day', y='total_bill', whis=[5, 95])
plt.title("Boxplot with Quartile Range Specified")
plt.show()
```

```python
# Example 8: Boxplot with symmetrical outliers
sns.boxplot(data=tips, x='day', y='total_bill', sym='b')
plt.title("Boxplot with Symmetrical Outliers")
plt.show()
```

```python
# Example 9: Boxplot with asymmetrical outliers
sns.boxplot(data=tips, x='day', y='total_bill', sym=[">", "<"])
plt.title("Boxplot with Asymmetrical Outliers")
plt.show()
```

```python
# Example 10: Boxplot with custom fliersize
sns.boxplot(data=tips, x='day', y='total_bill', fliersize=8)
plt.title("Boxplot with Custom Fliersize")
plt.show()
```

```python
# Example 11: Boxplot with vertical grid lines
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title("Boxplot with Vertical Grid Lines")
plt.grid(axis='y')
plt.show()
```

```python
# Example 12: Boxplot with horizontal grid lines
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title("Boxplot with Horizontal Grid Lines")
plt.grid(axis='x')
plt.show()
```

```python
# Example 13: Boxplot with whisker lines
sns.boxplot(data=tips, x='day', y='total_bill', showcaps=False)
plt.title("Boxplot with Whisker Lines")
plt.show()
```

```python
# Example 14: Boxplot with box outlines
sns.boxplot(data=tips, x='day', y='total_bill', showbox=False)
plt.title("Boxplot with Box Outlines")
plt.show()
```

```python
# Example 15: Boxplot with median lines
sns.boxplot(data=tips, x='day', y='total_bill', showmedians=True)
plt.title("Boxplot with Median Lines")
plt.show()
```

```python
# Example 16: Boxplot with mean points
sns.boxplot(data=tips, x='day', y='total_bill', showmeans=True)
plt.title("Boxplot with Mean Points")
plt.show()
```

```python
# Example 17: Boxplot with mean point color
sns.boxplot(data=tips, x='day', y='total_bill', showmeans=True, meanprops={"marker":"o", "markerfacecolor":"red"})
plt.title("Boxplot with Mean Point Color")
plt.show()
```

```python
# Example 18: Boxplot with whisker width
sns.boxplot(data=tips, x='day', y='total_bill', whiskerprops={'linewidth': 3})
plt.title("Boxplot with Whisker Width")
plt.show()
```

```python
# Example 19: Boxplot with cap width
sns.boxplot(data=tips, x='day', y='total_bill', capprops={'linewidth': 2})
plt.title("Boxplot with Cap Width")
plt.show()
```

```python
# Example 20: Boxplot with box width
sns.boxplot(data=tips, x='day', y='total_bill', boxprops={'linewidth': 2})
plt.title("Boxplot with Box Width")
plt.show()
```
