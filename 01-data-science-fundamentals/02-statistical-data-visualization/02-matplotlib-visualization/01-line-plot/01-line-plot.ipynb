```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for demonstration
np.random.seed(42)
x = np.linspace(0, 10, 30)
y = np.sin(x) + np.random.normal(0, 0.2, 30)

# Create a Line Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Sin(x) + Noise', color='blue', marker='o', linestyle='-', linewidth=2)

# Customize the plot
plt.title('Line Plot with Scatter Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Example 1: Change Marker Color and Size
plt.scatter(x, y, label='Data Points', color='red', marker='^', s=50, alpha=0.7)
```

```python
# Example 2: Change Marker Edge Color and Width
plt.scatter(x, y, label='Data Points', color='green', marker='s', edgecolors='black', linewidth=1.5)
```

```python
# Example 3: Vary Marker Size Based on a Parameter
sizes = np.abs(y) * 50
plt.scatter(x, y, label='Data Points', color='purple', marker='o', s=sizes)
```

```python
# Example 4: Add Color Map for Markers
colors = np.arange(len(x))
plt.scatter(x, y, label='Data Points', c=colors, cmap='viridis', marker='D', edgecolors='black')
```

```python
# Example 5: Connect Points with Lines
plt.scatter(x, y, label='Data Points', color='orange', marker='*', linestyle='dashed', linewidth=2)
```

```python
# Example 6: Highlight Specific Points
highlighted_indices = [5, 12, 20]
plt.scatter(x, y, label='Data Points', color='blue', marker='o')
plt.scatter(x[highlighted_indices], y[highlighted_indices], color='red', marker='o', label='Highlighted Points')
```

```python
# Example 7: Log Scale for Axis
plt.scatter(x, y, label='Data Points', color='cyan', marker='s')
plt.xscale('log')
```

```python
# Show legend
plt.legend()
```

```python
# Example 8: Multiple Lines on the Same Plot
y2 = np.cos(x) + np.random.normal(0, 0.2, 30)
plt.plot(x, y, label='Sin(x) + Noise', color='blue', marker='o', linestyle='-', linewidth=2)
plt.plot(x, y2, label='Cos(x) + Noise', color='red', marker='s', linestyle='--', linewidth=2)
```

```python
# Example 9: Change Line Style and Width
plt.plot(x, y, label='Sin(x) + Noise', color='green', marker='^', linestyle='-.', linewidth=3)
```

```python
# Example 10: Plot with Different Markers
plt.plot(x, y, label='Sin(x) + Noise', color='purple', marker='o', linestyle='-', linewidth=2)
plt.plot(x, y2, label='Cos(x) + Noise', color='orange', marker='s', linestyle='-', linewidth=2)
```

```python
# Example 11: Highlight Specific Range on the Plot
highlighted_range = (3, 7)
plt.plot(x, y, label='Sin(x) + Noise', color='cyan', marker='o', linestyle='-', linewidth=2)
plt.fill_between(x, y, where=((x >= highlighted_range[0]) & (x <= highlighted_range[1])), color='gray', alpha=0.3, label='Highlighted Range')
```

```python
# Example 12: Customizing Ticks and Tick Labels
plt.plot(x, y, label='Sin(x) + Noise', color='brown', marker='o', linestyle='-', linewidth=2)
plt.xticks(np.arange(0, 11, step=2))
plt.yticks(np.arange(-1.5, 2, step=0.5))
```

```python
# Example 13: Adding Annotations
plt.plot(x, y, label='Sin(x) + Noise', color='pink', marker='o', linestyle='-', linewidth=2)
plt.annotate('Maximum', xy=(np.argmax(y), np.max(y)), xytext=(8, 1), arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10)
```

```python
# Example 14: Log Scale for Y-axis
plt.plot(x, y, label='Sin(x) + Noise', color='magenta', marker='o', linestyle='-', linewidth=2)
plt.yscale('log')
```

```python
# Example 15: Change Background Color
plt.plot(x, y, label='Sin(x) + Noise', color='gray', marker='o', linestyle='-', linewidth=2)
plt.axhline(0, color='black', linewidth=0.5)  # Add a horizontal line at y=0
plt.axvline(8, color='red', linestyle='--', linewidth=1.5, label='Vertical Line at x=8')
plt.legend()
plt.gca().set_facecolor('lightyellow')
```
