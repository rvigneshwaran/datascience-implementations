```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
```

```python
# Example 1: Bessel Functions
x = np.linspace(0, 10, 100)
J0 = sp.j0(x)
J1 = sp.j1(x)
J2 = sp.jn(2, x)  # J_n(x) for n=2
plt.plot(x, J0, label='J0(x)')
plt.plot(x, J1, label='J1(x)')
plt.plot(x, J2, label='J2(x)')
plt.xlabel('x')
plt.ylabel('Bessel Functions')
plt.legend()
plt.title('Bessel Functions of the First Kind')
plt.grid(True)
plt.show()
```

```python
# Example 2: Modified Bessel Functions
x = np.linspace(0, 10, 100)
I0 = sp.i0(x)
I1 = sp.i1(x)
I2 = sp.iv(2, x)  # I_n(x) for n=2
plt.plot(x, I0, label='I0(x)')
plt.plot(x, I1, label='I1(x)')
plt.plot(x, I2, label='I2(x)')
plt.xlabel('x')
plt.ylabel('Modified Bessel Functions')
plt.legend()
plt.title('Modified Bessel Functions of the First Kind')
plt.grid(True)
plt.show()
```

```python
# Example 3: Spherical Bessel Functions
x = np.linspace(0, 10, 100)
j0 = sp.spherical_jn(0, x)  # j_n(x) for n=0
j1 = sp.spherical_jn(1, x)  # j_n(x) for n=1
j2 = sp.spherical_jn(2, x)  # j_n(x) for n=2
plt.plot(x, j0, label='j0(x)')
plt.plot(x, j1, label='j1(x)')
plt.plot(x, j2, label='j2(x)')
plt.xlabel('x')
plt.ylabel('Spherical Bessel Functions')
plt.legend()
plt.title('Spherical Bessel Functions')
plt.grid(True)
plt.show()
```

```python
# Example 4: Gamma Function
x = np.linspace(0.1, 5, 100)
gamma_x = sp.gamma(x)
plt.plot(x, gamma_x)
plt.xlabel('x')
plt.ylabel('Gamma Function')
plt.title('Gamma Function')
plt.grid(True)
plt.show()
```

```python
# Example 5: Digamma Function
x = np.linspace(0.1, 5, 100)
digamma_x = sp.digamma(x)
plt.plot(x, digamma_x)
plt.xlabel('x')
plt.ylabel('Digamma Function')
plt.title('Digamma Function')
plt.grid(True)
plt.show()
```

```python
# Example 6: Erf Function
x = np.linspace(-3, 3, 100)
erf_x = sp.erf(x)
plt.plot(x, erf_x)
plt.xlabel('x')
plt.ylabel('Erf Function')
plt.title('Error Function (Erf)')
plt.grid(True)
plt.show()
```

```python
# Example 7: Fresnel Functions
x = np.linspace(0, 5, 100)
s, c = sp.fresnel(x)
plt.plot(x, s, label='S(x)')
plt.plot(x, c, label='C(x)')
plt.xlabel('x')
plt.ylabel('Fresnel Functions')
plt.legend()
plt.title('Fresnel Functions')
plt.grid(True)
plt.show()
```

```python
# Example 8: Exponential Integral
x = np.linspace(0.1, 5, 100)
ei_x = sp.expi(x)
plt.plot(x, ei_x)
plt.xlabel('x')
plt.ylabel('Exponential Integral')
plt.title('Exponential Integral')
plt.grid(True)
plt.show()
```

```python
# Example 9: Airy Functions
x = np.linspace(-10, 10, 100)
Ai_x, Bi_x = sp.airy(x)
plt.plot(x, Ai_x, label='Ai(x)')
plt.plot(x, Bi_x, label='Bi(x)')
plt.xlabel('x')
plt.ylabel('Airy Functions')
plt.legend()
plt.title('Airy Functions')
plt.grid(True)
plt.show()
```

```python
# Example 10: Bessel Functions Zeros
n = 0
zeros_J0 = sp.jn_zeros(n, 10)
print(f"Zeros of J0(x) for n={n}: {zeros_J0}")
```
