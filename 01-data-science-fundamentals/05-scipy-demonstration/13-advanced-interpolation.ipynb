```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
# Example 1: Linear Interpolation with interp1d
x = np.linspace(0, 10, 11)
y = np.sin(x)
linear_interp = interpolate.interp1d(x, y, kind='linear')
x_new = np.linspace(0, 10, 101)
y_new = linear_interp(x_new)
plt.plot(x, y, 'o', label='Original Data')
plt.plot(x_new, y_new, '-', label='Linear Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Interpolation with interp1d')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 2: Cubic Spline Interpolation with splrep and splev
x = np.linspace(0, 10, 11)
y = np.sin(x)
tck = interpolate.splrep(x, y)
x_new = np.linspace(0, 10, 101)
y_new = interpolate.splev(x_new, tck)
plt.plot(x, y, 'o', label='Original Data')
plt.plot(x_new, y_new, '-', label='Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Spline Interpolation with splrep and splev')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 3: B-spline Interpolation with splrep and splev
x = np.linspace(0, 10, 11)
y = np.sin(x)
tck = interpolate.splrep(x, y, k=3)
x_new = np.linspace(0, 10, 101)
y_new = interpolate.splev(x_new, tck)
plt.plot(x, y, 'o', label='Original Data')
plt.plot(x_new, y_new, '-', label='B-spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.title('B-spline Interpolation with splrep and splev')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 5: Radial Basis Function (RBF) Interpolation with Rbf
x = np.linspace(0, 10, 11)
y = np.sin(x)
rbf_interp = interpolate.Rbf(x, y)
x_new = np.linspace(0, 10, 101)
y_new = rbf_interp(x_new)
plt.plot(x, y, 'o', label='Original Data')
plt.plot(x_new, y_new, '-', label='RBF Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Radial Basis Function (RBF) Interpolation with Rbf')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 6: Multivariate (2D) Interpolation with griddata
x = np.linspace(0, 5, 6)
y = np.linspace(0, 5, 6)
z = np.array([
    [20, 22, 24, 26, 28, 30],
    [22, 24, 26, 28, 30, 32],
    [24, 26, 28, 30, 32, 34],
    [26, 28, 30, 32, 34, 36],
    [28, 30, 32, 34, 36, 38],
    [30, 32, 34, 36, 38, 40]
])
x_new = np.linspace(0, 5, 100)
y_new = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x_new, y_new)
points = np.column_stack((x.ravel(), y.ravel()))
values = z.ravel()
z_new = interpolate.griddata(points, values, (X, Y), method='linear')
plt.contourf(X, Y, z_new, levels=20, cmap='viridis')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Interpolation with griddata')
plt.grid(True)
plt.show()
```

```python
# Example 7: Bivariate (2D) Interpolation with RectBivariateSpline
x = np.linspace(0, 5, 6)
y = np.linspace(0, 5, 6)
z = np.array([
    [20, 22, 24, 26, 28, 30],
    [22, 24, 26, 28, 30, 32],
    [24, 26, 28, 30, 32, 34],
    [26, 28, 30, 32, 34, 36],
    [28, 30, 32, 34, 36, 38],
    [30, 32, 34, 36, 38, 40]
])
interp_2d = interpolate.RectBivariateSpline(x, y, z)
x_new = np.linspace(0, 5, 100)
y_new = np.linspace(0, 5, 100)
z_new = interp_2d(x_new, y_new, grid=False)
X, Y = np.meshgrid(x_new, y_new)
plt.contourf(X, Y, z_new, levels=20, cmap='coolwarm')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Interpolation with RectBivariateSpline')
plt.grid(True)
plt.show()
```

```python
# Example 8: Trivariate (3D) Interpolation with RegularGridInterpolator
x = np.linspace(0, 5, 6)
y = np.linspace(0, 5, 6)
z = np.linspace(0, 5, 6)
data = np.random.rand(6, 6, 6)
interp_3d = interpolate.RegularGridInterpolator((x, y, z), data)
x_new = np.linspace(0, 5, 100)
y_new = np.linspace(0, 5, 100)
z_new = np.linspace(0, 5, 100)
points = np.array(np.meshgrid(x_new, y_new, z_new)).T.reshape(-1, 3)
values = interp_3d(points)
X, Y, Z = np.meshgrid(x_new, y_new, z_new)
plt.contourf(X[:, 0, 0], Y[0, :, 0], Z[0, 0, :], values.reshape(100, 100, 100)[:, :, 50], levels=20, cmap='plasma')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('3D Interpolation with RegularGridInterpolator')
plt.grid(True)
plt.show()
```

```python
# Example 9: Unstructured Grid Interpolation with griddata
x = np.random.rand(20)
y = np.random.rand(20)
z = np.sin(x) + np.cos(y)
x_new = np.linspace(0, 1, 100)
y_new = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x_new, y_new)
points = np.column_stack((x, y))
values = z
z_new = interpolate.griddata(points, values, (X, Y), method='linear')
plt.contourf(X, Y, z_new, levels=20, cmap='coolwarm')
plt.colorbar()
plt.scatter(x, y, c=z, cmap='coolwarm', edgecolors='black', linewidths=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Unstructured Grid Interpolation with griddata')
plt.grid(True)
plt.show()
```
