```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.spatial as sps
```

```python
# Example 1: Nearest Neighbor Search
points = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
query_point = np.array([0, 0])

# Create a KDTree data structure
kdtree = sps.KDTree(points)

# Find the nearest neighbor to the query point
nearest_neighbor_dist, nearest_neighbor_idx = kdtree.query(query_point)

# Plot the points and the query point
plt.scatter(points[:, 0], points[:, 1], label='Points')
plt.scatter(query_point[0], query_point[1], color='red', label='Query Point')
plt.scatter(points[nearest_neighbor_idx, 0], points[nearest_neighbor_idx, 1], color='green', label='Nearest Neighbor')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Nearest Neighbor Search')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 2: Nearest Neighbor Search with Multiple Query Points
query_points = np.array([[1, 1], [4, 5], [6, 7]])

# Find the nearest neighbors to the query points
nearest_neighbor_dists, nearest_neighbor_indices = kdtree.query(query_points, k=2)

# Plot the points and the query points
plt.scatter(points[:, 0], points[:, 1], label='Points')
plt.scatter(query_points[:, 0], query_points[:, 1], color='red', label='Query Points')
plt.scatter(points[nearest_neighbor_indices[:, 0], 0], points[nearest_neighbor_indices[:, 0], 1], color='green', label='Nearest Neighbor 1')
plt.scatter(points[nearest_neighbor_indices[:, 1], 0], points[nearest_neighbor_indices[:, 1], 1], color='blue', label='Nearest Neighbor 2')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Nearest Neighbor Search with Multiple Query Points')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 3: Delaunay Triangulation
points = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# Create a Delaunay triangulation
tri = sps.Delaunay(points)

# Plot the triangulation
plt.triplot(points[:, 0], points[:, 1], tri.simplices)
plt.scatter(points[:, 0], points[:, 1], color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Delaunay Triangulation')
plt.grid(True)
plt.show()
```

```python
# Example 4: Convex Hull
points = np.array([[1, 1], [2, 2], [2, 3], [3, 2], [1, 5], [4, 2]])

# Compute the convex hull
hull = sps.ConvexHull(points)

# Plot the points and the convex hull
plt.plot(points[:, 0], points[:, 1], 'o')
for simplex in hull.simplices:
    plt.plot(points[simplex, 0], points[simplex, 1], 'k-')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Convex Hull')
plt.grid(True)
plt.show()
```

```python
# Example 5: Voronoi Diagram
points = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# Create a Voronoi diagram
vor = sps.Voronoi(points)

# Plot the Voronoi diagram
sps.voronoi_plot_2d(vor)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Voronoi Diagram')
plt.grid(True)
plt.show()
```

```python
# Example 6: Nearest Neighbor Clustering
points = np.random.rand(20, 2)  # Random points

# Create a KDTree data structure
kdtree = sps.KDTree(points)

# Perform nearest neighbor clustering with a maximum distance threshold
max_distance = 0.3
clusters = kdtree.query_ball_point(points, max_distance)

# Plot the points and the clusters
for cluster_indices in clusters:
    cluster_points = points[cluster_indices]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Nearest Neighbor Clustering')
plt.grid(True)
plt.show()
```

```python
# Example 7: K-Means Clustering
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate sample data
data, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Perform K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(data)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Plot the clustered data and centroids
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red', s=100, label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 8: Agglomerative Clustering
from sklearn.datasets import make_moons
from sklearn.cluster import AgglomerativeClustering

# Generate sample data
data, _ = make_moons(n_samples=200, noise=0.05, random_state=42)

# Perform Agglomerative clustering
agg_clustering = AgglomerativeClustering(n_clusters=2)
agg_clustering.fit(data)
labels = agg_clustering.labels_

# Plot the clustered data
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Agglomerative Clustering')
plt.grid(True)
plt.show()
```
