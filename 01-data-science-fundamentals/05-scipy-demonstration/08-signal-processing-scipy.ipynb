```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# Example 1: Generating a Sine Wave
t = np.linspace(0, 1, 1000, endpoint=False)
f = 5  # Frequency (in Hz)
A = 1  # Amplitude
sine_wave = A * np.sin(2 * np.pi * f * t)
plt.plot(t, sine_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.grid(True)
plt.show()
```

```python
# Example 2: Generating a Square Wave
square_wave = signal.square(2 * np.pi * 5 * t)
plt.plot(t, square_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Square Wave')
plt.grid(True)
plt.show()
```

```python
# Example 3: Generating a Sawtooth Wave
sawtooth_wave = signal.sawtooth(2 * np.pi * 5 * t)
plt.plot(t, sawtooth_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sawtooth Wave')
plt.grid(True)
plt.show()
```

```python
# Example 4: Generating a Chirp Signal
t_chirp = np.linspace(0, 10, 1000)
chirp_signal = signal.chirp(t_chirp, f0=1, f1=6, t1=10, method='linear')
plt.plot(t_chirp, chirp_signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Chirp Signal')
plt.grid(True)
plt.show()
```

```python
# Example 5: Generating a Gaussian Window
window = signal.gaussian(M=31, std=5)
plt.plot(window)
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.title('Gaussian Window')
plt.grid(True)
plt.show()
```

```python
# Example 6: Applying a Low-Pass Filter
t_signal = np.linspace(0, 1, 1000, endpoint=False)
signal_noise = 0.5 * np.sin(2 * np.pi * 10 * t_signal) + 0.5 * np.sin(2 * np.pi * 20 * t_signal)
signal_filtered = signal.lfilter([1], [1, -0.9], signal_noise)
plt.plot(t_signal, signal_noise, label='Noisy Signal')
plt.plot(t_signal, signal_filtered, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Low-Pass Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 7: Applying a High-Pass Filter
signal_filtered = signal.lfilter([1, -0.9], [1], signal_noise)
plt.plot(t_signal, signal_noise, label='Noisy Signal')
plt.plot(t_signal, signal_filtered, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('High-Pass Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 8: Applying a Band-Pass Filter
signal_filtered = signal.lfilter([1, 0, -1], [1, -0.9], signal_noise)
plt.plot(t_signal, signal_noise, label='Noisy Signal')
plt.plot(t_signal, signal_filtered, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Band-Pass Filter')
plt.legend()
plt.grid(True)
plt.show()
```
