```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# Example 1: Generating a Sine Wave
t = np.linspace(0, 1, 1000, endpoint=False)
f = 5  # Frequency (in Hz)
A = 1  # Amplitude
sine_wave = A * np.sin(2 * np.pi * f * t)
plt.plot(t, sine_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.grid(True)
plt.show()
```

```python
# Example 2: Generating a Square Wave
square_wave = signal.square(2 * np.pi * 5 * t)
plt.plot(t, square_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Square Wave')
plt.grid(True)
plt.show()
```

```python
# Example 3: Generating a Sawtooth Wave
sawtooth_wave = signal.sawtooth(2 * np.pi * 5 * t)
plt.plot(t, sawtooth_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sawtooth Wave')
plt.grid(True)
plt.show()
```

```python
# Example 4: Generating a Chirp Signal
t_chirp = np.linspace(0, 10, 1000)
chirp_signal = signal.chirp(t_chirp, f0=1, f1=6, t1=10, method='linear')
plt.plot(t_chirp, chirp_signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Chirp Signal')
plt.grid(True)
plt.show()
```

```python
# Example 5: Generating a Gaussian Window
window = signal.gaussian(M=31, std=5)
plt.plot(window)
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.title('Gaussian Window')
plt.grid(True)
plt.show()
```

```python
# Example 6: Applying a Low-Pass Filter
t_signal = np.linspace(0, 1, 1000, endpoint=False)
signal_noise = 0.5 * np.sin(2 * np.pi * 10 * t_signal) + 0.5 * np.sin(2 * np.pi * 20 * t_signal)
signal_filtered = signal.lfilter([1], [1, -0.9], signal_noise)
plt.plot(t_signal, signal_noise, label='Noisy Signal')
plt.plot(t_signal, signal_filtered, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Low-Pass Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 7: Applying a High-Pass Filter
signal_filtered = signal.lfilter([1, -0.9], [1], signal_noise)
plt.plot(t_signal, signal_noise, label='Noisy Signal')
plt.plot(t_signal, signal_filtered, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('High-Pass Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 8: Applying a Band-Pass Filter
signal_filtered = signal.lfilter([1, 0, -1], [1, -0.9], signal_noise)
plt.plot(t_signal, signal_noise, label='Noisy Signal')
plt.plot(t_signal, signal_filtered, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Band-Pass Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 9: Applying a Band-Stop Filter
signal_filtered = signal.lfilter([1, -0.9], [1, 0, -1], signal_noise)
plt.plot(t_signal, signal_noise, label='Noisy Signal')
plt.plot(t_signal, signal_filtered, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Band-Stop Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 10: Fast Fourier Transform (FFT)
t_fft = np.linspace(0, 1, 1000, endpoint=False)
signal_fft = 2 * np.sin(2 * np.pi * 5 * t_fft) + 3 * np.cos(2 * np.pi * 10 * t_fft)
signal_spectrum = np.abs(np.fft.fft(signal_fft))
freqs = np.fft.fftfreq(len(t_fft), d=1/1000)
plt.plot(freqs, signal_spectrum)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Fast Fourier Transform (FFT)')
plt.grid(True)
plt.show()
```

```python
# Example 11: Short-Time Fourier Transform (STFT)
t_stft = np.linspace(0, 10, 1000)
signal_stft = 2 * np.sin(2 * np.pi * 2 * t_stft) + 3 * np.cos(2 * np.pi * 5 * t_stft)
frequencies, times, stft_data = signal.stft(signal_stft, fs=100, nperseg=100)
plt.pcolormesh(times, frequencies, np.abs(stft_data), shading='gouraud')
plt.colorbar(label='Magnitude')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Short-Time Fourier Transform (STFT)')
plt.grid(True)
plt.show()
```

```python
# Example 12: Inverse Short-Time Fourier Transform (ISTFT)
t_istft, signal_istft = signal.istft(stft_data, fs=100, nperseg=100)
plt.plot(t_stft, signal_stft, label='Original Signal')
plt.plot(t_istft, signal_istft, label='Reconstructed Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Inverse Short-Time Fourier Transform (ISTFT)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 13: Wavelet Transform
t_wavelet = np.linspace(0, 1, 1000, endpoint=False)
signal_wavelet = np.sin(2 * np.pi * 10 * t_wavelet) + np.cos(2 * np.pi * 20 * t_wavelet)
coeffs, freqs = signal.cwt(signal_wavelet, signal.ricker, widths=np.arange(1, 31))
plt.imshow(np.abs(coeffs), aspect='auto', extent=[0, 1, 1, 31], cmap='jet')
plt.colorbar(label='Magnitude')
plt.xlabel('Time (s)')
plt.ylabel('Scale (Width of Ricker Wavelet)')
plt.title('Wavelet Transform')
plt.grid(True)
plt.show()
```

```python
# Example 14: Convolution
signal_conv = np.convolve(signal_wavelet, window, mode='same')
plt.plot(t_wavelet, signal_wavelet, label='Original Signal')
plt.plot(t_wavelet, signal_conv, label='Convolved Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Convolution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 15: Correlation
signal_corr = signal.correlate(signal_wavelet, window, mode='same')
plt.plot(t_wavelet, signal_wavelet, label='Original Signal')
plt.plot(t_wavelet, signal_corr, label='Correlated Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Correlation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 16: Resample Signal
t_original = np.linspace(0, 10, 1000)
signal_original = np.sin(2 * np.pi * 5 * t_original)
t_resampled = np.linspace(0, 10, 500)
signal_resampled = signal.resample(signal_original, 500)
plt.plot(t_original, signal_original, label='Original Signal')
plt.plot(t_resampled, signal_resampled, label='Resampled Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Resample Signal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 17: Decimate Signal
t_decimate = np.linspace(0, 10, 1000)
signal_decimate = np.sin(2 * np.pi * 5 * t_decimate)
signal_decimated = signal.decimate(signal_decimate, q=5)
t_decimated = t_decimate[::5]
plt.plot(t_decimate, signal_decimate, label='Original Signal')
plt.plot(t_decimated, signal_decimated, label='Decimated Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Decimate Signal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 18: Hilbert Transform
t_hilbert = np.linspace(0, 1, 1000)
signal_hilbert = np.sin(2 * np.pi * 5 * t_hilbert)
hilbert_transform = signal.hilbert(signal_hilbert)
plt.plot(t_hilbert, signal_hilbert, label='Original Signal')
plt.plot(t_hilbert, np.abs(hilbert_transform), label='Hilbert Transform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Hilbert Transform')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 19: Envelope of Signal
envelope = signal.hilbert(signal_hilbert)
plt.plot(t_hilbert, signal_hilbert, label='Original Signal')
plt.plot(t_hilbert, np.abs(envelope), label='Envelope')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Envelope of Signal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 20: Group Delay
b, a = signal.butter(4, 0.2)
w, gd = signal.group_delay((b, a))
plt.plot(w, gd)
plt.xlabel('Frequency (rad/sample)')
plt.ylabel('Group Delay (samples)')
plt.title('Group Delay')
plt.grid(True)
plt.show()
```

```python
# Example 21: Transfer Function from Frequency Response
frequency = np.arange(1, 1000)
response = 1 / (1j * 2 * np.pi * frequency + 1)
num, den = signal.freqresp((response, [1]))
plt.plot(frequency, np.abs(num / den))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Transfer Function from Frequency Response')
plt.grid(True)
plt.show()
```

```python
# Example 22: Spectrogram
t_spectrogram = np.linspace(0, 10, 1000)
signal_spectrogram = 2 * np.sin(2 * np.pi * 2 * t_spectrogram) + 3 * np.cos(2 * np.pi * 5 * t_spectrogram)
frequencies, times, Sxx = signal.spectrogram(signal_spectrogram, fs=100, nperseg=100)
plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud')
plt.colorbar(label='Power Spectral Density (dB/Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Spectrogram')
plt.grid(True)
plt.show()
```

```python
# Example 23: Find Peaks in Signal
signal_peaks = np.array([0, 1, 2, 1, 0, -1, -2, -1, 0, 1, 2, 1, 0])
peaks, _ = signal.find_peaks(signal_peaks)
plt.plot(signal_peaks)
plt.plot(peaks, signal_peaks[peaks], 'x')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.title('Find Peaks in Signal')
plt.grid(True)
plt.show()
```

```python
# Example 24: Find Peaks Above Threshold
threshold = 0.5
peaks, _ = signal.find_peaks(signal_peaks, height=threshold)
plt.plot(signal_peaks)
plt.plot(peaks, signal_peaks[peaks], 'x')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.title('Find Peaks Above Threshold')
plt.grid(True)
plt.show()
```

```python
# Example 25: Find Peaks with Prominence
peaks, _ = signal.find_peaks(signal_peaks, prominence=1)
plt.plot(signal_peaks)
plt.plot(peaks, signal_peaks[peaks], 'x')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.title('Find Peaks with Prominence')
plt.grid(True)
plt.show()
```
