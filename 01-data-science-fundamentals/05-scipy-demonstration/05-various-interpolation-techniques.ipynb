```python
# Import necessary libraries
import numpy as np
import scipy.interpolate as spi
import matplotlib.pyplot as plt
```

```python
# Example 1: Linear Interpolation with interp1d
# Given data points (x, y) representing the power of a house over time
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([100, 200, 400, 600, 800, 1000])

# Create a linear interpolation function
linear_interp = spi.interp1d(x, y, kind='linear')

# Generate new data points and plot the interpolated curve
x_new = np.linspace(0, 5, 100)
y_new = linear_interp(x_new)
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x_new, y_new, '-', label='Linear Interpolation')
plt.xlabel('Time')
plt.ylabel('Power')
plt.title('Linear Interpolation of House Power over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 2: Cubic Spline Interpolation with interp1d
# Given data points (x, y) representing the power of a house over time
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([100, 200, 400, 600, 800, 1000])

# Create a cubic spline interpolation function
cubic_interp = spi.interp1d(x, y, kind='cubic')

# Generate new data points and plot the interpolated curve
x_new = np.linspace(0, 5, 100)
y_new = cubic_interp(x_new)
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x_new, y_new, '-', label='Cubic Spline Interpolation')
plt.xlabel('Time')
plt.ylabel('Power')
plt.title('Cubic Spline Interpolation of House Power over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 3: Barycentric Interpolation with barycentric_interpolate
# Given data points (x, y) representing the population of a city over time
x = np.array([1900, 1910, 1920, 1930, 1940, 1950])
y = np.array([5000, 7000, 9000, 12000, 15000, 18000])

# Generate new data points and perform barycentric interpolation
x_new = np.linspace(1900, 1950, 100)
y_new = spi.barycentric_interpolate(x, y, x_new)

# Plot the interpolated curve
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x_new, y_new, '-', label='Barycentric Interpolation')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Barycentric Interpolation of City Population over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 4: 2D Interpolation with interp2d
# Given data points (x, y, z) representing the power of a house over time and resources
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([100, 200, 300, 400, 500, 600])
z = np.array([
    [150, 250, 350, 450, 550, 650],
    [160, 260, 360, 460, 560, 660],
    [170, 270, 370, 470, 570, 670],
    [180, 280, 380, 480, 580, 680],
    [190, 290, 390, 490, 590, 690],
    [200, 300, 400, 500, 600, 700]
])

# Create a 2D interpolation function
interp_2d = spi.interp2d(x, y, z, kind='linear')

# Generate new data points and plot the interpolated surface
x_new = np.linspace(0, 5, 100)
y_new = np.linspace(100, 600, 100)
z_new = interp_2d(x_new, y_new)
X, Y = np.meshgrid(x_new, y_new)
plt.contourf(X, Y, z_new, levels=20, cmap='viridis')
plt.colorbar()
plt.xlabel('Time')
plt.ylabel('Power')
plt.title('2D Linear Interpolation of House Power over Time and Resources')
plt.grid(True)
plt.show()
```

```python
# Example 5: Spline Interpolation with InterpolatedUnivariateSpline
# Given data points (x, y) representing the power of a house over time
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([100, 200, 400, 600, 800, 1000])

# Create a spline interpolation function
spline_interp = spi.InterpolatedUnivariateSpline(x, y)

# Generate new data points and plot the interpolated curve
x_new = np.linspace(0, 5, 100)
y_new = spline_interp(x_new)
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x_new, y_new, '-', label='Spline Interpolation')
plt.xlabel('Time')
plt.ylabel('Power')
plt.title('Spline Interpolation of House Power over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 6: Nearest Neighbor Interpolation with interp1d
# Given data points (x, y) representing the power of a house over time
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([100, 200, 400, 600, 800, 1000])

# Create a nearest neighbor interpolation function
nearest_interp = spi.interp1d(x, y, kind='nearest')

# Generate new data points and plot the interpolated curve
x_new = np.linspace(0, 5, 100)
y_new = nearest_interp(x_new)
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x_new, y_new, '-', label='Nearest Neighbor Interpolation')
plt.xlabel('Time')
plt.ylabel('Power')
plt.title('Nearest Neighbor Interpolation of House Power over Time')
plt.legend()
plt.grid(True)
plt.show()
```
