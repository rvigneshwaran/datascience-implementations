```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
# Example 1: Loading and Displaying an Image
image = plt.imread('example_image.jpg')
plt.imshow(image)
plt.axis('off')
plt.title('Original Image')
plt.show()
```

```python
# Example 2: Convert Image to Grayscale
image_gray = np.dot(image[...,:3], [0.2989, 0.5870, 0.1140])
plt.imshow(image_gray, cmap='gray')
plt.axis('off')
plt.title('Grayscale Image')
plt.show()
```

```python
# Example 3: Image Flipping
image_flipped = np.flipud(image)
plt.imshow(image_flipped)
plt.axis('off')
plt.title('Flipped Image')
plt.show()
```

```python
# Example 4: Image Rotation
image_rotated = ndimage.rotate(image, angle=45, reshape=False)
plt.imshow(image_rotated)
plt.axis('off')
plt.title('Rotated Image')
plt.show()
```

```python
# Example 5: Image Resizing
image_resized = ndimage.zoom(image, zoom=0.5)
plt.imshow(image_resized)
plt.axis('off')
plt.title('Resized Image')
plt.show()
```

```python
# Example 6: Image Blurring (Gaussian Filter)
image_blurred = ndimage.gaussian_filter(image, sigma=2)
plt.imshow(image_blurred)
plt.axis('off')
plt.title('Blurred Image')
plt.show()
```

```python
# Example 7: Image Sharpening (Unsharp Masking)
image_sharpened = image + 1.5 * (image - image_blurred)
plt.imshow(image_sharpened)
plt.axis('off')
plt.title('Sharpened Image')
plt.show()
```

```python
# Example 8: Edge Detection (Sobel Filter)
image_edge = ndimage.sobel(image_gray)
plt.imshow(image_edge, cmap='gray')
plt.axis('off')
plt.title('Edge Detection')
plt.show()
```

```python
# Example 9: Erosion (Morphological Operation)
image_binary = image_gray > 100  # Thresholding to create binary image
image_eroded = ndimage.binary_erosion(image_binary)
plt.imshow(image_eroded, cmap='gray')
plt.axis('off')
plt.title('Erosion')
plt.show()
```

```python
# Example 10: Dilation (Morphological Operation)
image_dilated = ndimage.binary_dilation(image_binary)
plt.imshow(image_dilated, cmap='gray')
plt.axis('off')
plt.title('Dilation')
plt.show()
```
