```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, expon, chi2, t, poisson, binom, gamma, kendalltau, mannwhitneyu
```

```python
# Example 1: Normal Distribution
x = np.linspace(-5, 5, 1000)
mean, std = 0, 1
pdf = norm.pdf(x, mean, std)
plt.plot(x, pdf, label='Normal Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Normal Distribution (mean=0, std=1)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 2: Uniform Distribution
x = np.linspace(0, 1, 1000)
pdf = uniform.pdf(x, loc=0, scale=1)
plt.plot(x, pdf, label='Uniform Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Uniform Distribution (loc=0, scale=1)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 3: Exponential Distribution
x = np.linspace(0, 5, 1000)
pdf = expon.pdf(x, scale=1)
plt.plot(x, pdf, label='Exponential Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution (scale=1)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 4: Chi-Squared Distribution
x = np.linspace(0, 10, 1000)
pdf = chi2.pdf(x, df=3)
plt.plot(x, pdf, label='Chi-Squared Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Chi-Squared Distribution (df=3)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 5: t-Distribution
x = np.linspace(-5, 5, 1000)
pdf = t.pdf(x, df=5)
plt.plot(x, pdf, label='t-Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('t-Distribution (df=5)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 6: Poisson Distribution
x = np.arange(0, 10)
pmf = poisson.pmf(x, mu=2)
plt.stem(x, pmf, label='Poisson Distribution')
plt.xlabel('X')
plt.ylabel('Probability Mass')
plt.title('Poisson Distribution (mu=2)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 7: Binomial Distribution
x = np.arange(0, 11)
pmf = binom.pmf(x, n=10, p=0.5)
plt.stem(x, pmf, label='Binomial Distribution')
plt.xlabel('X')
plt.ylabel('Probability Mass')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 8: Gamma Distribution
x = np.linspace(0, 5, 1000)
pdf = gamma.pdf(x, a=2)
plt.plot(x, pdf, label='Gamma Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Gamma Distribution (a=2)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 9: Random Number Generation from Normal Distribution
mean, std = 0, 1
samples = norm.rvs(loc=mean, scale=std, size=1000)
plt.hist(samples, bins=20, density=True, alpha=0.6, color='blue')
plt.plot(x, pdf, label='Normal Distribution')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Random Number Generation from Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 17: Kendall's Tau Test
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 2, 1, 4, 5])
tau, p_value = kendalltau(x, y)
print(f"Kendall's Tau: {tau}, p-value: {p_value}")
```

```python
# Example 18: Mann-Whitney U Test
x = np.random.rand(50)
y = np.random.rand(60)
statistic, p_value = mannwhitneyu(x, y)
print(f"Mann-Whitney U Statistic: {statistic}, p-value: {p_value}")
```

```python
# Example 24: Spectrogram
fs = 1000  # Sampling frequency
t = np.linspace(0, 1, fs)
x = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 20 * t) + np.random.randn(fs)
plt.specgram(x, Fs=fs, NFFT=256, noverlap=128)
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Spectrogram')
plt.colorbar(label='Intensity (dB)')
plt.grid(True)
plt.show()
```
