```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# Example 1: Creating a Sine Wave Signal
t = np.linspace(0, 1, 1000)
freq = 5
amplitude = 2
sine_wave = amplitude * np.sin(2 * np.pi * freq * t)
plt.plot(t, sine_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave Signal')
plt.grid(True)
plt.show()
```

```python
# Example 2: Creating a Square Wave Signal
square_wave = signal.square(2 * np.pi * freq * t)
plt.plot(t, square_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Square Wave Signal')
plt.grid(True)
plt.show()
```

```python
# Example 3: Creating a Sawtooth Wave Signal
sawtooth_wave = signal.sawtooth(2 * np.pi * freq * t)
plt.plot(t, sawtooth_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sawtooth Wave Signal')
plt.grid(True)
plt.show()
```

```python
# Example 4: Designing a Low-pass Butterworth Filter
b, a = signal.butter(4, 0.2, 'low')
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(len(t))
filtered_x = signal.lfilter(b, a, x)
plt.plot(t, x, label='Original Signal')
plt.plot(t, filtered_x, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Low-pass Butterworth Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 5: Designing a High-pass Butterworth Filter
b, a = signal.butter(4, 0.2, 'high')
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(len(t))
filtered_x = signal.lfilter(b, a, x)
plt.plot(t, x, label='Original Signal')
plt.plot(t, filtered_x, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('High-pass Butterworth Filter')
plt.legend()
plt.grid(True)
plt.show()
```
