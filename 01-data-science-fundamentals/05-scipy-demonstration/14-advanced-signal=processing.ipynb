```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# Example 1: Creating a Sine Wave Signal
t = np.linspace(0, 1, 1000)
freq = 5
amplitude = 2
sine_wave = amplitude * np.sin(2 * np.pi * freq * t)
plt.plot(t, sine_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave Signal')
plt.grid(True)
plt.show()
```

```python
# Example 2: Creating a Square Wave Signal
square_wave = signal.square(2 * np.pi * freq * t)
plt.plot(t, square_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Square Wave Signal')
plt.grid(True)
plt.show()
```

```python
# Example 3: Creating a Sawtooth Wave Signal
sawtooth_wave = signal.sawtooth(2 * np.pi * freq * t)
plt.plot(t, sawtooth_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sawtooth Wave Signal')
plt.grid(True)
plt.show()
```

```python
# Example 4: Designing a Low-pass Butterworth Filter
b, a = signal.butter(4, 0.2, 'low')
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(len(t))
filtered_x = signal.lfilter(b, a, x)
plt.plot(t, x, label='Original Signal')
plt.plot(t, filtered_x, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Low-pass Butterworth Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 5: Designing a High-pass Butterworth Filter
b, a = signal.butter(4, 0.2, 'high')
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(len(t))
filtered_x = signal.lfilter(b, a, x)
plt.plot(t, x, label='Original Signal')
plt.plot(t, filtered_x, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('High-pass Butterworth Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 6: Designing a Band-pass Butterworth Filter
b, a = signal.butter(4, [0.2, 0.8], 'band')
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(len(t))
filtered_x = signal.lfilter(b, a, x)
plt.plot(t, x, label='Original Signal')
plt.plot(t, filtered_x, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Band-pass Butterworth Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 7: Designing a Band-stop Butterworth Filter
b, a = signal.butter(4, [0.2, 0.8], 'bandstop')
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(len(t))
filtered_x = signal.lfilter(b, a, x)
plt.plot(t, x, label='Original Signal')
plt.plot(t, filtered_x, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Band-stop Butterworth Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 8: Calculating the Power Spectral Density (PSD) of a Signal
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(len(t))
frequencies, psd = signal.welch(x)
plt.semilogy(frequencies, psd)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power/Frequency (dB/Hz)')
plt.title('Power Spectral Density (PSD)')
plt.grid(True)
plt.show()
```

```python
# Example 9: Applying a Hann Window to a Signal
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t)
windowed_x = signal.hann(len(x)) * x
plt.plot(t, x, label='Original Signal')
plt.plot(t, windowed_x, label='Windowed Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Hann Window')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 12: Applying a Kaiser Window to a Signal
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t)
windowed_x = signal.kaiser(len(x), beta=14) * x
plt.plot(t, x, label='Original Signal')
plt.plot(t, windowed_x, label='Windowed Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Kaiser Window')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 14: Applying a Tukey Window to a Signal
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t)
windowed_x = signal.tukey(len(x), alpha=0.3) * x
plt.plot(t, x, label='Original Signal')
plt.plot(t, windowed_x, label='Windowed Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Tukey Window')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 15: Calculating the Short-Time Fourier Transform (STFT) of a Signal
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 10 * t)
frequencies, times, spectrogram = signal.stft(x)
plt.pcolormesh(times, frequencies, np.abs(spectrogram), shading='gouraud')
plt.colorbar(label='Amplitude')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Short-Time Fourier Transform (STFT)')
plt.grid(True)
plt.show()
```

```python
# Example 16: Applying a Savitzky-Golay Filter to a Signal
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + 0.1 * np.random.randn(len(t))
filtered_x = signal.savgol_filter(x, window_length=51, polyorder=3)
plt.plot(t, x, label='Original Signal')
plt.plot(t, filtered_x, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Savitzky-Golay Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 17: Applying a Butterworth Band-pass Filter to a Signal
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + 0.1 * np.random.randn(len(t))
b, a = signal.butter(4, [0.2, 0.8], 'band')
filtered_x = signal.lfilter(b, a, x)
plt.plot(t, x, label='Original Signal')
plt.plot(t, filtered_x, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Butterworth Band-pass Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 18: Finding Peaks in a Signal using FindPeaksCWT
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(len(t))
peaks, properties = signal.find_peaks_cwt(x, widths=np.arange(1, 10))
plt.plot(t, x, label='Signal with Peaks')
plt.plot(t[peaks], x[peaks], 'ro', label='Detected Peaks')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Finding Peaks in a Signal using FindPeaksCWT')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 19: Applying a Chebyshev Type I Band-stop Filter to a Signal
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + 0.1 * np.random.randn(len(t))
b, a = signal.cheby1(4, 1, [0.2, 0.8], 'bandstop')
filtered_x = signal.lfilter(b, a, x)
plt.plot(t, x, label='Original Signal')
plt.plot(t, filtered_x, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Chebyshev Type I Band-stop Filter')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example 20: Designing a FIR Filter using the Parks-McClellan Algorithm
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + 0.1 * np.random.randn(len(t))
bands = [0, 0.2, 0.3, 0.5, 0.6, 0.8, 0.9, 1]
desired = [0, 1, 0, 0]
b = signal.firls(101, bands, desired)
filtered_x = signal.lfilter(b, 1, x)
plt.plot(t, x, label='Original Signal')
plt.plot(t, filtered_x, label='Filtered Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('FIR Filter Design using the Parks-McClellan Algorithm')
plt.legend()
plt.grid(True)
plt.show()
```
