```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Example 1: Normal Distribution
data_normal = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data_normal, bins=30, density=True, alpha=0.6, color='blue')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution')
plt.show()
```

```python
# Example 2: Exponential Distribution
data_exponential = np.random.exponential(scale=1, size=1000)
plt.hist(data_exponential, bins=30, density=True, alpha=0.6, color='green')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Exponential Distribution')
plt.show()
```

```python
# Example 3: Uniform Distribution
data_uniform = np.random.uniform(low=0, high=1, size=1000)
plt.hist(data_uniform, bins=30, density=True, alpha=0.6, color='orange')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Uniform Distribution')
plt.show()
```

```python
# Example 4: Bernoulli Distribution
data_bernoulli = np.random.binomial(n=1, p=0.5, size=1000)
plt.hist(data_bernoulli, bins=2, density=True, alpha=0.6, color='red')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Bernoulli Distribution')
plt.show()
```

```python
# Example 5: T-Test
data_group1 = np.random.normal(loc=0, scale=1, size=100)
data_group2 = np.random.normal(loc=1, scale=1, size=100)
t_statistic, p_value = stats.ttest_ind(data_group1, data_group2)
print("T-Test:")
print("T-Statistic =", t_statistic)
print("P-Value =", p_value)
```

```python
# Example 6: Chi-Square Test
data_observed = np.array([[10, 20, 30], [15, 25, 35]])
chi2_stat, p_value, dof, expected = stats.chi2_contingency(data_observed)
print("Chi-Square Test:")
print("Chi2 Statistic =", chi2_stat)
print("P-Value =", p_value)
print("Degrees of Freedom =", dof)
print("Expected Frequencies:")
print(expected)
```

```python
# Example 7: Kolmogorov-Smirnov Test
data_sample = np.random.normal(loc=0, scale=1, size=100)
ks_statistic, ks_p_value = stats.kstest(data_sample, 'norm')
print("Kolmogorov-Smirnov Test:")
print("KS Statistic =", ks_statistic)
print("KS P-Value =", ks_p_value)
```

```python
# Example 8: F-Test
data_group1 = np.random.normal(loc=0, scale=1, size=100)
data_group2 = np.random.normal(loc=0, scale=2, size=100)
f_statistic, f_p_value = stats.f_oneway(data_group1, data_group2)
print("F-Test:")
print("F-Statistic =", f_statistic)
print("P-Value =", f_p_value)
```

```python
# Example 9: Mann-Whitney U Test
data_group1 = np.random.normal(loc=0, scale=1, size=100)
data_group2 = np.random.normal(loc=1, scale=1, size=100)
u_statistic, u_p_value = stats.mannwhitneyu(data_group1, data_group2)
print("Mann-Whitney U Test:")
print("U-Statistic =", u_statistic)
print("P-Value =", u_p_value)
```

```python
# Example 10: Pearson Correlation Coefficient
data_x = np.random.normal(loc=0, scale=1, size=100)
data_y = np.random.normal(loc=0, scale=1, size=100)
correlation_coefficient, _ = stats.pearsonr(data_x, data_y)
print("Pearson Correlation Coefficient:")
print("Correlation Coefficient =", correlation_coefficient)
```

```python
# Example 11: Spearman Rank Correlation Coefficient
data_x = np.random.normal(loc=0, scale=1, size=100)
data_y = np.random.normal(loc=0, scale=1, size=100)
correlation_coefficient, _ = stats.spearmanr(data_x, data_y)
print("Spearman Rank Correlation Coefficient:")
print("Correlation Coefficient =", correlation_coefficient)
```

```python
# Example 12: Kendall Rank Correlation Coefficient
data_x = np.random.normal(loc=0, scale=1, size=100)
data_y = np.random.normal(loc=0, scale=1, size=100)
correlation_coefficient, _ = stats.kendalltau(data_x, data_y)
print("Kendall Rank Correlation Coefficient:")
print("Correlation Coefficient =", correlation_coefficient)
```

```python
# Example 13: Descriptive Statistics
data = np.random.normal(loc=0, scale=1, size=100)
mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_dev = np.std(data)
print("Descriptive Statistics:")
print("Mean =", mean)
print("Median =", median)
print("Variance =", variance)
print("Standard Deviation =", std_dev)
```
